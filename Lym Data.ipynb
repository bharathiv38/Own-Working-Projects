### Splitting into X and Y

```python
X = cancer.drop(columns='lymphatics',axis=1)
Y = cancer['lymphatics']
```

### splitting data into train test 

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test =train_test_split(X,Y,test_size=0.1,random_state=0)
```

### Model Evaluation

```python
#accuracy on training data
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
lg_reg = LogisticRegression()
```

```python
lg_reg.fit(X_train,Y_train)
```

```python
X_train_prediction = lg_reg.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print("The accuracy score on training data :",training_data_accuracy)
```

```python
X_test_prediction = lg_reg.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print("The accuracy score on test data :",test_data_accuracy)
```
