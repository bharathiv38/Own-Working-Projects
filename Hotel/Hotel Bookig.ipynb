# Highest number of passengers

```python
print('The highest range of average daily rate is: {}'.format(house['adr'].max()),'\n','The name of that person is: {}'.format(house.iloc[house['adr'].idxmax()]['name']))
```

```python
print('The average range of adr is: {}'.format(round(house['adr'].mean(),2)))
```

```python
house['total_stays']= house['stays_in_weekend_nights']+house['stays_in_week_nights']
print('The average range of stays is: {}'.format(round(house['total_stays'].mean(),2)))
```

```python
special_request= house[house['total_of_special_requests']== 5]
special_request[['total_of_special_requests', 'name', 'email']]
```

```python
def last_name(name):
    return(name.split()[1])

house['last_name']= house['name'].apply(last_name)
house['last_name'].value_counts().iloc[0:5]
```

```python
house['total_children']= house['babies']+house['children']
house.sort_values('total_children', ascending=False).iloc[0:5][['total_children', 'name']]
```

```python
def area_code(code):
    return(code.split('-')[0])
house['area_code']= house['phone-number'].apply(area_code)
area_grouped= house.groupby('area_code').count()
area_grouped.sort_values('hotel', ascending=False).iloc[0:3]['hotel']
```
