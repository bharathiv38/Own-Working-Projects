## Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
dataset.iloc[:,1] = labelencoder.fit_transform(dataset.iloc[:,1].values)
dataset.head()
```

We did label encoding as we have to make the dataset look with numbers

## Train Test & Split

```python
X = dataset.iloc[:,2:].values
Y = dataset.iloc[:,1].values
```

```python
# as this is our target column we are rechecking
print(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

## Model Training

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
lr = LogisticRegression()
```

```python
#training the model
lr.fit(X_train,Y_train)
```

## Model Evaluation

```python
# Accuracy on the training dataset
X_train_predict = lr.predict(X_train)
training_data_accuracy =accuracy_score(Y_train,X_train_predict)
```

```python
print("Accuracy score of the training dataset is : ",training_data_accuracy)
```

```python
#accuracy on the test dataset
X_test_predict = lr.predict(X_test)
testing_data_accuracy = accuracy_score(Y_test,X_test_predict)
```

```python
print("Accuracy score of the testing dataset is : ",testing_data_accuracy)
```
